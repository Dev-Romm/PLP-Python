```python

import pandas as pd
import matplotlib.pyplot as plt


try:
    # Assuming the CSV file is named 'week_7_data.csv' and is in the same directory as this script
    # Load the data from the CSV file
    data = pd.read_csv('week_7_data.csv')
    print(data.head())
    print(data.describe())

    # Groupings on Gender column and numericals using Age
    print("\nMEAN AGE:", data['Age'].groupby(data['Gender']).mean())
    print("\nMAX AGE:", data['Age'].groupby(data['Gender']).max())
    print("\nMIN AGE:", data['Age'].groupby(data['Gender']).min())
except FileNotFoundError:
    print("The file 'week_7_data.csv' was not found. Please ensure it is in the same directory as this script.")
    exit(1)
except pd.errors.EmptyDataError:
    print("The file 'week_7_data.csv' is empty. Please check the file.")
    exit(1)
except pd.errors.ParserError:
    print("Error parsing the file 'week_7_data.csv'. Please check the file format.")
    exit(1)
except Exception as e:
    print(f"An unexpected error occurred: {e}")
    exit(1)
    

# Plotting a line graph with Age on the x-axis and Name on the y-axis
plt.plot(data['Age'], data['Name'], marker='o', linestyle='-')
plt.xlabel('Age')
plt.ylabel('Name')
plt.grid(True)
plt.title('A line plot of Name against Age', fontweight='bold')
plt.xticks(rotation=45)
plt.show()


# Plotting a bar graph with Name on the x-axis and Age on the y-axis
plt.bar(data['Name'], data['Age'], color='blue')
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('A bar plot of Age against Name', fontweight='bold')
plt.show()

# Plotting a histogram for Age
plt.hist(data['Age'], bins=10, color='green', alpha=0.7, edgecolor='black')
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('A histogram for Age Frequency', fontweight='bold')
plt.show()

# Plotting a scatter plot with Age on the x-axis and Name on the y-axis
plt.scatter(data['Age'], data['Name'], color='red', alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Name')
plt.title('A scatter plot of Name against Age', fontweight='bold')
plt.grid(True)
plt.show()

# Plotting a pie chart for the distribution of Age
plt.pie(data['Age'], labels=data['Name'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')  
plt.title('Age Distribution', fontweight='bold')
plt.show()



```
